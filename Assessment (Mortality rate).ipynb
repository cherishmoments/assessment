```python
# import required modules
import pandas as pd
import numpy as np
import sqlite3
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
adult_mortality = pd.read_csv("/Users/jinyee/Downloads/archive/Adult mortality.csv")
adult_mortality
```

```python
adult_mortality_new = adult_mortality.rename(columns={"Adult mortality rate (probability of dying between 15 and 60 years per 1000 population)": "adult_mortality_both_gender", "Unnamed: 0": "Country", "Unnamed: 1": "Year", "Adult mortality rate (probability of dying between 15 and 60 years per 1000 population).1":"adult_mortality_male", "Adult mortality rate (probability of dying between 15 and 60 years per 1000 population).2": "adult_mortality_female" })
adult_mortality_new = adult_mortality_new.drop([0])
adult_mortality_new
```

```python
adult_mortality_new["Year"] = adult_mortality_new["Year"].astype(int)
adult_mortality_new["adult_mortality_both_gender"] = adult_mortality_new["adult_mortality_both_gender"].astype(int)
adult_mortality_new["adult_mortality_male"] = adult_mortality_new["adult_mortality_male"].astype(int)
adult_mortality_new["adult_mortality_female"] = adult_mortality_new["adult_mortality_female"].astype(int)
```

```python
deaths = pd.read_csv("/Users/jinyee/Downloads/archive/Number of deaths (thousands).csv")
deaths
```

```python
deaths_new = deaths.rename(columns={"Number of infant deaths (thousands).1": "infant_deaths_(thousands)_Male", "Number of infant deaths (thousands).2": "infant_deaths_(thousands)_Female", "Number of infant deaths (thousands)": "infant_deaths_(thousands)_both_gender" })
deaths_new = deaths_new.rename(columns={"Number of under-five deaths (thousands)": "under_five_deaths_(thousands)_both_gender", "Number of under-five deaths (thousands).1": "under_five_deaths_(thousands)_male", "Number of under-five deaths (thousands).2": "under_five_deaths_(thousands)_female"})
deaths_new = deaths_new.rename(columns={"Unnamed: 0": "Country", "Unnamed: 1": "Year"})
deaths_new = deaths_new.rename(columns={"Number of neonatal deaths (thousands)": "neonatal_deaths_(thousands)_both_gender"})
deaths_new = deaths_new.drop([0])
deaths_new
```

```python
deaths_new["Year"] = deaths_new["Year"].astype(int)
deaths_new["under_five_deaths_(thousands)_both_gender"] = deaths_new["under_five_deaths_(thousands)_both_gender"].astype(int)
deaths_new["under_five_deaths_(thousands)_male"] = deaths_new["under_five_deaths_(thousands)_male"].astype(int)
deaths_new["under_five_deaths_(thousands)_female"] = deaths_new["under_five_deaths_(thousands)_female"].astype(int)
deaths_new["infant_deaths_(thousands)_both_gender"] = deaths_new["infant_deaths_(thousands)_both_gender"].astype(int)
deaths_new["infant_deaths_(thousands)_Male"] = deaths_new["infant_deaths_(thousands)_Male"].astype(int)
deaths_new["infant_deaths_(thousands)_Female"] = deaths_new["infant_deaths_(thousands)_Female"].astype(int)
deaths_new["neonatal_deaths_(thousands)_both_gender"] = deaths_new["neonatal_deaths_(thousands)_both_gender"].astype(int)
```

```python
conn = sqlite3.connect('Mydatabase.db')
```

```python
adult_mortality_new.to_sql('adult_mortality_new', conn, index=False, if_exists='replace')
```

```python
deaths_new.to_sql('deaths_new', conn, index=False, if_exists='replace')
```

```python
children_mortality= pd.read_csv("/Users/jinyee/Downloads/archive/Probability of dying per 1000 live births.csv")
children_mortality
```

```python
children_mortality_new = children_mortality.rename(columns={"Infant mortality rate (probability of dying between birth and age 1 per 1000 live births)": "infant__mortality_both_gender", "Infant mortality rate (probability of dying between birth and age 1 per 1000 live births).1": "infant_mortality_male", "Infant mortality rate (probability of dying between birth and age 1 per 1000 live births).2": "infant_mortality_female" })
children_mortality_new = children_mortality_new.rename(columns={"Under-five mortality rate (probability of dying by age 5 per 1000 live births)": "under_five_deaths_mortality_both_gender", "Under-five mortality rate (probability of dying by age 5 per 1000 live births).1": "under_five_mortality_male", "Under-five mortality rate (probability of dying by age 5 per 1000 live births).2": "under_five_mortality_female"})
children_mortality_new = children_mortality_new.rename(columns={"Unnamed: 0": "Country", "Unnamed: 1": "Year"})
children_mortality_new = children_mortality_new.rename(columns={"Neonatal mortality rate (per 1000 live births)": "neonatal_mortality_both_gender"})
children_mortality_new = children_mortality_new.drop([0])
children_mortality_new
```

```python
children_mortality_new["Year"] = children_mortality_new["Year"].astype(int)
children_mortality_new["infant__mortality_both_gender"] = children_mortality_new["infant__mortality_both_gender"].astype(float)
children_mortality_new["infant_mortality_male"] = children_mortality_new["infant_mortality_male"].astype(float)
children_mortality_new["infant_mortality_female"] = children_mortality_new["infant_mortality_female"].astype(float)
children_mortality_new["under_five_deaths_mortality_both_gender"] = children_mortality_new["under_five_deaths_mortality_both_gender"].astype(float)
children_mortality_new["under_five_mortality_male"] = children_mortality_new["under_five_mortality_male"].astype(float)
children_mortality_new["under_five_mortality_female"] = children_mortality_new["under_five_mortality_female"].astype(float)
children_mortality_new["neonatal_mortality_both_gender"] = children_mortality_new["neonatal_mortality_both_gender"].astype(float)
```

```python
children_mortality_new.to_sql('children_mortality', conn, index=False, if_exists='replace')
```

```python
children_mortality = pd.read_sql_query('SELECT * FROM children_mortality', conn)
```

```python
year_2016 = pd.read_sql_query('SELECT * FROM children_mortality WHERE (Year == 2016)', conn)
year_2016
```

```python
year_2016.to_sql('children_mortality_2016', conn, index=False, if_exists='replace')
```

```python
top5_year_2016 = pd.read_sql_query('SELECT * FROM children_mortality_2016 GROUP BY children_mortality_2016.Year, children_mortality_2016.Country ORDER BY children_mortality_2016.infant__mortality_both_gender DESC LIMIT 5;', conn)  
top5_year_2016
```

```python
top5_year_2016.to_sql('children_mortality_2016_top5', conn, index=False, if_exists='replace')
```

```python
pd.read_sql_query("select * from sqlite_master where type='table';", conn)
```

```python
children_mortality_top_5 = pd.read_sql_query('SELECT * FROM children_mortality WHERE children_mortality.Country IN (SELECT children_mortality_2016_top5.Country FROM children_mortality_2016_top5);', conn) 
children_mortality_top_5
```

```python
fig3 = px.line(children_mortality_top_5, x='Year', y='infant__mortality_both_gender', color='Country',
               labels = {'infant__mortality_both_gender': 'Infant Morality rate (Both genders) (probability of dying between birth and age 1 per 1000 live births))'},markers=True, title="Infant Mortality Rate for the Top 5 Highest Rate countries in 2016")
fig3.update_yaxes(
        title_text = 'Infant Morality rate(Both genders)(prob of dying between birth and age 1 per 1000 live births)',
        title_font = {"size": 9})
fig3.show()
```
